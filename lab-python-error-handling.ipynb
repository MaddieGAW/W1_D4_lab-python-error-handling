````python
products = ["tshirt", "mug", "hat", "book", "keychain"]

def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            except ValueError:
                print("Invalid input. Please enter a valid quantity.")
    return inventory

inventory = initialize_inventory(products)
````

3. Modify the get_customer_orders function to include error handling.

If the user enters an invalid number of orders (e.g., a negative value or a non-numeric value), display an error message and ask them to re-enter the number of orders. If the user enters an invalid product name (e.g., a product name that is not in the inventory), or that doesn't have stock available, display an error message and ask them to re-enter the product name. Hint: you will need to pass inventory as a parameter Use a try-except block to handle the error and continue prompting the user until a valid product name is entered.

````python
def get_customer_orders(inventory):
    customer_orders_set = set()
    while True:
        try:
            number_customer_orders = int(input("Enter the number of customer orders: "))
            if number_customer_orders <= 0:
                raise ValueError("Invalid number of orders. Please enter a positive integer.")
            break
        except ValueError:
            print("Error: Please enter a valid number.")

    for _ in range(number_customer_orders):
        while True:
            order = input("Enter the name of a product that a customer wants to order: ")
            if order not in inventory:
                print("We do not sell this product.")
            elif inventory[order] == 0:
                print("This product is out of stock.")
            else:
                customer_orders_set.add(order)
                break
    try:
        return customer_orders_set
    finally:
        print("Thank you for your order. Order received! Order:", customer_orders_set)

customer_orders = get_customer_orders(inventory)
````

````python
def update_inventory(customer_orders, inventory):
  updated_inventory = {product: inventory[product] - 1 if product in customer_orders else inventory[product] for product in inventory}
  updated_inventory = {product: count for product, count in updated_inventory.items() if count > 0}
  return updated_inventory

updated_inventory = update_inventory(customer_orders, inventory)

update_inventory(customer_orders, inventory)
````

````python
def calculate_order_statistics(customer_orders, products):
  total_products_ordered = len(customer_orders)
  inventory_total_items = sum(updated_inventory.values())
  percentage_of_unique_products_ordered = (total_products_ordered / inventory_total_items) * 100
  order_statistics = (total_products_ordered, percentage_of_unique_products_ordered)
  return order_statistics

order_statistics = calculate_order_statistics(customer_orders, products)
````

````python
def print_order_statistics(order_statistics):
    print("Order Statistics:")
    print("The number of products ordered is", order_statistics[0])
    print("The percentage of unique products ordered is", order_statistics[1])

print_order_statistics(order_statistics)
````

````python
def print_updated_inventory(updated_inventory):
  print("Updated Inventory: ")
  for product, quantity in updated_inventory.items():
    print(f"{product} : {quantity}")

print_updated_inventory(updated_inventory)
````

2. Modify the calculate_total_price function to include error handling.

If the user enters an invalid price (e.g., a negative value or a non-numeric value), display an error message and ask them to re-enter the price for that product.
Use a try-except block to handle the error and continue prompting the user until a valid price is entered.

````python
def total_price_customer_order(customer_orders):
    while True:
        prices = []
        valid_input = True
        for product in customer_orders:
            price_input = input(f"Enter the price of {product}: ")
            if not price_input.isdigit():
                print("Invalid input. Please enter a valid price.")
                valid_input = False
                break
            price = int(price_input)
            prices.append(price)
        if valid_input:
            total_price = sum(prices)
            total_price = print("Total Price:", total_price)
            return total_price
````
